```python
import os
import shutil
import random

def copy_random_files(src_folder, dest_folder, num_files_to_copy):
    # Get list of all files in the source folder
    all_files = [f for f in os.listdir(src_folder) if os.path.isfile(os.path.join(src_folder, f))]
    
    # Check if there are enough files to copy
    if len(all_files) < num_files_to_copy:
        print(f"Warning: Only {len(all_files)} files found. Adjusting to copy {len(all_files)} files.")
        num_files_to_copy = len(all_files)

    # Randomly select files to copy
    files_to_copy = random.sample(all_files, num_files_to_copy)

    # Copy files to the destination folder
    for file_name in files_to_copy:
        full_file_path = os.path.join(src_folder, file_name)
        shutil.copy(full_file_path, dest_folder)
        print(f"Copied: {file_name}")

    print(f"Successfully copied {num_files_to_copy} files to {dest_folder}")

# Example usage
src_folder = r'C:\Users\rishi\Downloads\AWS2024'
dest_folder = r'C:\Users\rishi\Downloads\AWSPost'
num_files_to_copy = 100  # Number of random files you want to copy

copy_random_files(src_folder, dest_folder, num_files_to_copy)
```
