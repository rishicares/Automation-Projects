```python
import re
def comvert_emp_length(emp_length):
    #Remove "years" and any whitespace characters
    emp_length = re.sun(r'\s*years*', '', emp_length)

    #Convert "<1" to 0 and "10+" to 15
    if emp_length=='<1':
        return 0
    elif emp_length=="10+":
        return 15
    else: return int(emp_length)

    #Apply the conversion function to the emp_length column
    df['emp_length'] = df['emp_length'].apply(convert_emp_length)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

#select features for clustering
features=df[['emp_length', 'grade', 'home_ownership', 'loan_status', 'purpose',
       'term', 'verification_status', 'annual_income', 'dti', 'installment',
       'int_rate', 'loan_amount', 'total_acc', 'total_payment']]

wcss =[]

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=42)
    kmeans.fit(features_scaled)
    wcss.append(kmeans.inertia_)

#plot elbow curve
plt.plot(range(1,11),wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(features_scaled)

df['cluster'] = kmeans.labels_
df
```

```python
#silhouette score
from sklearn.metrics import silhouette_score

avg = silhouette_score(features_scaled,kmeans.labels_)
print(f'silhouette score:{avg:.4f}')
```
