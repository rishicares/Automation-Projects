```python
df['Installs']=df['Installs'].str.replace(r'\+','',regex=True)
df['Installs']=df['Installs'].str.replace(r'\,','',regex=True)
df['Installs']=df['Installs'].str.replace('Free','0.0')
df['Installs']=df['Installs'].astype(float)
```
